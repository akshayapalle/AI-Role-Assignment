# 4. Working with Video Captures

```python
cap = cv2.VideoCapture(os.path.join('data','videos','AIvideo.mp4'))
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    print(frame_idx)
```

```python
#Establish capture
cap = cv2.VideoCapture(os.path.join('data','videos','AIvideo.mp4'))
#Loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    #Read frame
    ret, frame = cap.read()
    
    #if you want the video in gray color, can use the below code otherwise can avoid the below line
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    #Show image
    cv2.imshow('Video Player', gray)#if you want gray video, after 'Video Player' write 'gray' otherwise write 'frame'
    
    #Breaking out of the lopp
    if cv2.waitKey(10) & 0xFF == ord('q'):# clicking 'q' will shutdown the window
        break
    
#Close down everything
cap.release()
cv2.destroyAllWindows()
```

# 5. Writing out Videos

```python
#Establish capture
cap = cv2.VideoCapture(os.path.join('data','videos','AIvideo.mp4'))

#Properties
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = cap.get(cv2.CAP_PROP_FPS)
#Video Writer
video_writer = cv2.VideoWriter(os.path.join('data','output','output.avi'), cv2.VideoWriter_fourcc('P','I','M','1'),fps,(width, height), isColor=False)


#Loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    #Read frame
    ret, frame = cap.read()
    
    #if you want the video in gray color, can use the below code otherwise can avoid the below line
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    #Show image
    cv2.imshow('Video Player', gray)#if you want gray video, after 'Video Player' write 'gray' otherwise write 'frame'
    #write out frame
    video_writer.write(gray)
    
    
    #Breaking out of the lopp
    if cv2.waitKey(10) & 0xFF == ord('q'):# clicking 'q' will shutdown the window
        break
    
#Close down everything
cap.release()
cv2.destroyAllWindows()

#Release video writer
video_writer.release()
```
